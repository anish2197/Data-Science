## Problem Statement :

The Fashion MNIST Dataset contains 28x28 grayscale 70,000 images, each associated with a label from the following 10 classes:

- T-shirt
- Trouser
- Pullover
- Dress
- Coat
- Sandal
- Shirt
- Sneaker
- Bag
- Ankle boot

Each image is 28 pixels in height and 28 pixels in width, for a total of 784 pixels in total. Each pixel has a single pixel-value associated with it ranging from 0 to 255, indicating its brightness or darkness.
