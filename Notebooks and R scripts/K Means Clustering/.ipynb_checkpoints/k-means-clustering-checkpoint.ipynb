```python
import pandas as pd
import numpy as np
from copy import deepcopy
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
#Importing dataset
data = pd.read_csv("xclara.csv")
print(data.shape)
data.head()
```

```python
#Getting X and Y values
x = data["V1"].values
y = data["V2"].values
arr = np.array(list(zip(x,y)))
plt.figure(figsize=(10,7))
plt.scatter(x,y,c="black",s=7)
```

```python
#Function to find Euclidean Distance

def dist(x,y,ax=1):
    return np.linalg.norm(x-y,axis=ax)
```

```python
#Defining cluster number and centroids
k = 3

centroids_x = np.random.randint(0,np.max(x)-20,size = k)

centroids_y = np.random.randint(0,np.max(x)-20,size = k)

centroids = np.array(list(zip(centroids_x,centroids_y)), dtype = np.float32)
centroids
```

```python
#Plotting data with centroids

plt.figure(figsize=(10,7))
plt.scatter(x,y,c="black", s = 7)
plt.scatter(centroids_x,centroids_y,marker="*",s = 100,c ="red")
```

```python
#K means algorithm

centroids_old = np.zeros(centroids.shape)
clusters = np.zeros(len(arr))

error = dist(centroids,centroids_old,None)

len1 = len(arr)

while error != 0:
    
    for i in range(len1):
        
        distances = dist(arr[i],centroids)
        cluster = np.argmin(distances)
        clusters[i] = cluster
        
    centroids_old = deepcopy(centroids)
        
    for i in range(k):
        points = [arr[j] for j in range(len(arr)) if clusters[j] == i ]
        centroids[i] = np.mean(points, axis = 0)
        
    error = (centroids,centroids_old)

    
colors = ["r","g","b"]

plt.figure(figsize=(10,7))
sub = plt.subplots()

for i in range(k):
    points = [arr[j] for j in range(len(arr)) if clusters[j] == i]
    sub.scatter(points[:,0],points[:,1], s = 7, c = colors[i])

            
        
        
        
```
